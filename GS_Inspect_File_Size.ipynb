```python
import os
import shutil

# 검색할 디렉토리와 이동할 디렉토리 설정
source_directory = r"D:\SamsungSTF\Data\GSmbiz\BMS_Data"
destination_directory = r"D:\SamsungSTF\Data\GSmbiz\Zero_Size_Files"

# 이동할 디렉토리가 없으면 생성
if not os.path.exists(destination_directory):
    os.makedirs(destination_directory)

# 디렉토리 순회
for root, dirs, files in os.walk(source_directory):
    for file in files:
        file_path = os.path.join(root, file)
        # 파일 크기가 0KB인지 확인
        if os.path.getsize(file_path) == 0:
            # 파일 이동
            shutil.move(file_path, destination_directory)
            print(f'Moved: {file_path} to {destination_directory}')

print("모든 0KB 파일이 이동되었습니다.")
```
